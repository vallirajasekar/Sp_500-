## Adding Additional Features 

## Simple Moving Average

The Simple Moving Average is the average of a specific number of prices over a defined period.

For example, let's calculate the 3-day SMA:

Day 2 SMA = (50 + 55 + 52) / 3 = 52.33
Day 3 SMA = (55 + 52 + 58) / 3 = 55


```python
sp500_1['SMA30'] = sp500_1['Close'].rolling(30).mean()
 
# removing all the NULL values using
# dropna() method
sp500_1.dropna(inplace=True)
 
# printing Dataframe
sp500_1.head()
```

## Cummulative Moving Average 

Cumulative Moving Average (CMA):

The Cumulative Moving Average is the average of all prices up to that point.

To calculate the 3-day CMA:

Day 1 CMA = 50
Day 2 CMA = (50 + 55) / 2 = 52.5
Day 3 CMA = (50 + 55 + 52) / 3 = 52.33

```python
#sp500_1 = sp500_1['Close'].to_frame()
 
# calculating cumulative moving
# average using .expanding().mean()
sp500_1['CMA30'] = sp500_1['Close'].expanding().mean()
 
# printing Dataframe
sp500_1
```

## Exponential Moving Average 
The Exponential Moving Average gives more weight to recent prices and discounts older prices.

To calculate the 3-day EMA:

Start with the 3-day SMA on the first day: EMA1 = 52.33
EMA2 = (55 - EMA1) * (2 / (3 + 1)) + EMA1 = 53.83
EMA3 = (52 - EMA2) * (2 / (3 + 1)) + EMA2 = 52.57
...

```python
sp500_1['EWMA30'] = sp500_1['Close'].ewm(span=30).mean()
 
```
